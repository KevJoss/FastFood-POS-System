## Prepare DataFrame
First we transform the Database of names (xsl file) to csv_file to make it easier to work with.

```python
import pandas as pd

# Ruta del archivo .xlsx
xlsx_file = r'C:\Users\NW\Documents\QUINTO SEMESTRE\Advanced Programming\Final Project\POS_System\DataBase\CODIGOS ABRIL 2025.xlsx'

# Ruta del archivo .csv de salida
csv_file = r'C:\Users\NW\Documents\QUINTO SEMESTRE\Advanced Programming\Final Project\POS_System\DataBase\CODIGOS.csv'

# Leer el archivo .xlsx
df = pd.read_excel(xlsx_file)

# Guardar como .csv
df.to_csv(csv_file, index=False)
```

```python
df
```

Change the name of columns, and delete the NaN values

```python
df = df.rename(columns={
    'PRODUCTO': 'PRODUCTO',
    'PRECIO': 'PRECIO',
    'PRODUCTO.1' : 'PRODUCTO',
    'PRECIO.1' : 'PRECIO'})
dataBase = df.drop(df.columns[2], axis=1)
```

```python
dataBase = dataBase.fillna('')
dataBase
```

Now we must to split the DataFrame into different DataFrames according to the type of products of restaurant

```python
dataBase1 = dataBase.iloc[:, :2]
dataBase2 = dataBase.iloc[:, 2:]
dataBase1
```

```python
dataBase2
```

We make the DataFrame for storing Dishes with Name : Price

```python
Dishes = {
    'PRODUCTO': [i for i in dataBase1.iloc[:6, dataBase1.columns.get_loc('PRODUCTO')]],
    'PRECIO' : [i for i in dataBase1.iloc[:6, dataBase1.columns.get_loc('PRECIO')]]
}
newDishes = pd.DataFrame({
    'PRODUCTO': [i for i in dataBase2.iloc[:3, dataBase2.columns.get_loc('PRODUCTO')]],
    'PRECIO' : [i for i in dataBase2.iloc[:3, dataBase2.columns.get_loc('PRECIO')]]
})
Dishes = pd.DataFrame(Dishes)
Dishes = pd.concat([Dishes, newDishes], ignore_index=True)
print("\tDishes DataFrame")
Dishes
```

```python
products = []
for idx, row in Dishes.iterrows():
    if pd.notna(row['PRODUCTO']):
        products.append({
            "name" : row['PRODUCTO'],
            "price" : row['PRECIO']
        })
import json

with open('Dishes.json', 'w') as dishes:
    json.dump({"Dishes": products}, dishes, ensure_ascii=False, indent=4)
```

We make the DataFrame for storing Combos with Name : Price

```python
Combos = {
    'PRODUCTO': [i for i in dataBase1.iloc[6:21, dataBase1.columns.get_loc('PRODUCTO')]],
    'PRECIO' : [i for i in dataBase1.iloc[6:21, dataBase1.columns.get_loc('PRECIO')]]
}
Combos = pd.DataFrame(Combos)
Combos = Combos.drop(12)
print("\tCombos DataFrame")
Combos
```

```python
products = []
for idx, row in Combos.iterrows():
    if pd.notna(row['PRODUCTO']):
        products.append({
            "name" : row['PRODUCTO'],
            "price" : row['PRECIO']
        })

with open('Combos.json', 'w') as combos:
    json.dump({"Combos": products}, combos, ensure_ascii=False, indent=4)
```
